```python
df = df.drop_duplicates()
```

```python
df
```

```python
# DROP DATE COLUMN WHERE BLANKS COLUMNS 

df['Order_Date'] = pd.to_datetime(df['Order_Date'],errors='coerce')

df = df.dropna(subset=['Order_Date'])
```

```python
#create new columns

df['Year'] = df['Order_Date'].dt.year
```

```python
df
```

```python
df['Months'] = df['Order_Date'].dt.month

df
```

```python
df['Profit'] = df ['Sales'] - df['Cost']
```

```python
df
```

```python
# For save data

df.to_csv('Sales_cleaned.csv', index= False)   # index = False for index column 
```
